```python
import matplotlib.pyplot as plt

# Inicio pregunta E
tiempo_inicial=0
tiempo_final=4.19
velocidad_inicial=5
diferencia=0.01

numero_particiones=int((tiempo_final-tiempo_inicial)/diferencia)

lista_tiempo_x=[tiempo_inicial]
lista_velocidad_y=[velocidad_inicial]


aceleracion = lambda v : -9.8*(0.1+(0.2-0.1)/(1+2*v))

for particion in range (numero_particiones):
    velocidad_inicial = velocidad_inicial + diferencia*aceleracion(velocidad_inicial)
    lista_velocidad_y.append(velocidad_inicial)
    tiempo_inicial = tiempo_inicial + diferencia
    lista_tiempo_x.append(tiempo_inicial)

# Fin pregunta E


coefD = 0.1
coefS = 0.2
VelocidadC = 0.5
Vm = 1.5

def acel2(v,coefD,coefS,VelocidadC,Vm ):
    Uv =  coefD + (coefS-coefD) /  (1  + (v/VelocidadC))
    return Uv*-9.8 



lista_acel12 =[]

for i in range(len(lista_velocidad_y)):
    velocidad_temporaria = lista_velocidad_y[i]
    aceleracion_temporaria = acel2(velocidad_temporaria,coefD,coefS,VelocidadC,Vm)
    lista_acel12.append(aceleracion_temporaria)


fig, gx = plt.subplots()
gx.plot(lista_tiempo_x, lista_acel12)
gx.set(xlabel='Tiempo (s)', ylabel='Aceleracion(m/s2)',
       title='Gráfica de la situación')
gx.grid()
```

```python

import matplotlib.pyplot as plt

tiempo_inicial=0
tiempo_final=5.22
velocidad_inicial=5

diferencia=0.01

numero_particiones=int((tiempo_final-tiempo_inicial)/diferencia)
lista_tiempo_x=[tiempo_inicial]
lista_velocidad_y=[velocidad_inicial]

def funcion_para_g(v):
    if v>1.5:
        ecuacion = -9.8*(0.125*((v/1.5)**(-1/2)))
        return ecuacion
    else:
        ecuacion = -9.8*(0.1+(0.2-0.1)/(1+2*v))
        return ecuacion


for i in range (numero_particiones):
    velocidad_inicial = velocidad_inicial + diferencia*funcion_para_g(velocidad_inicial)
    lista_velocidad_y.append(velocidad_inicial)
    
    
    tiempo_inicial = tiempo_inicial + diferencia
    lista_tiempo_x.append(tiempo_inicial)

######
coefD = 0.1
coefS = 0.2
VelocidadC = 0.5
Vm = 1.5
######


def acel3(v,coefD,coefS,VelocidadC,Vm):
    u_m = coefD +(coefS-coefD)/(1+(Vm/VelocidadC))
    if v > Vm:
        AV = -9.8*(u_m*((v/Vm)**(-0.5)))
    else:
        Uv= coefD+(coefS-coefD)/(1+(v/VelocidadC))
        AV=-9.8*Uv
    return AV
    



lista_acel3 = []
for i in range(len(lista_velocidad_y)):
    v= lista_velocidad_y[i]
    A_V=acel3(v,coefD,coefS,VelocidadC,Vm)
    lista_acel3.append(A_V)

    
    
fig, gx = plt.subplots()

gx.plot(lista_tiempo_x, lista_acel3)
gx.set(xlabel='Tiempo (s)', ylabel='Velocidad (m/s)',
       title='Gráfica de la situación')

gx.grid()
```
