```python
import matplotlib.pyplot as plt

# Inicio pregunta E
tiempo_inicial_2=0
tiempo_final_2=4.19
velocidad_inicial_2=5
diferencia_2=0.01

numero_particiones_2=int((tiempo_final_2-tiempo_inicial_2)/diferencia_2)

lista_tiempo_x_2=[tiempo_inicial_2]
lista_velocidad_y_2=[velocidad_inicial_2]


aceleracion = lambda v : -9.8*(0.1+(0.2-0.1)/(1+2*v))

for particion in range (numero_particiones_2):
    velocidad_inicial_2 = velocidad_inicial_2 + diferencia_2*aceleracion(velocidad_inicial_2)
    lista_velocidad_y_2.append(velocidad_inicial_2)
    tiempo_inicial_2 = tiempo_inicial_2 + diferencia_2
    lista_tiempo_x_2.append(tiempo_inicial_2)

# Fin pregunta E


coefD = 0.1
coefS = 0.2
VelocidadC = 0.5
Vm = 1.5

def acel2(v,coefD,coefS,VelocidadC,Vm ):
    Uv =  coefD + (coefS-coefD) /  (1  + (v/VelocidadC))
    return Uv*-9.8 



lista_acel2 =[]

for i in range(len(lista_velocidad_y_2)):
    velocidad_temporaria = lista_velocidad_y_2[i]
    aceleracion_temporaria = acel2(velocidad_temporaria,coefD,coefS,VelocidadC,Vm)
    lista_acel2.append(aceleracion_temporaria)


```

```python

tiempo_inicial_3=0
tiempo_final_3=5.22
velocidad_inicial_3=5

diferencia_3=0.01

numero_particiones_3=int((tiempo_final_3-tiempo_inicial_3)/diferencia_3)
lista_tiempo_x_3=[tiempo_inicial_3]
lista_velocidad_y_3=[velocidad_inicial_3]

def funcion_para_g(v):
    if v>1.5:
        ecuacion = -9.8*(0.125*((v/1.5)**(-1/2)))
        return ecuacion
    else:
        ecuacion = -9.8*(0.1+(0.2-0.1)/(1+2*v))
        return ecuacion


for i in range (numero_particiones_3):
    velocidad_inicial_3 = velocidad_inicial_3 + diferencia_3*funcion_para_g(velocidad_inicial_3)
    lista_velocidad_y_3.append(velocidad_inicial_3)
    
    
    tiempo_inicial_3 = tiempo_inicial_3 + diferencia_3
    lista_tiempo_x_3.append(tiempo_inicial_3)

######
coefD = 0.1
coefS = 0.2
VelocidadC = 0.5
Vm = 1.5
######


def acel3(v,coefD,coefS,VelocidadC,Vm):
    u_m = coefD +(coefS-coefD)/(1+(Vm/VelocidadC))
    if v > Vm:
        AV = -9.8*(u_m*((v/Vm)**(-0.5)))
    else:
        Uv= coefD+(coefS-coefD)/(1+(v/VelocidadC))
        AV=-9.8*Uv
    return AV
    



lista_acel3 = []
for i in range(len(lista_velocidad_y_3)):
    v= lista_velocidad_y_3[i]
    A_V=acel3(v,coefD,coefS,VelocidadC,Vm)
    lista_acel3.append(A_V)

```

```python
fig, gx = plt.subplots()

gx.plot(lista_tiempo_x_1, lista_acel1, color ="red")
gx.plot(lista_tiempo_x_2, lista_acel2, color="green")
gx.plot(lista_tiempo_x_3, lista_acel3, color ="blue")

gx.set(xlabel='Tiempo (s)', ylabel='Aceleracion (m/s)',
       title='Gráfica de la situación')

gx.grid()
```
