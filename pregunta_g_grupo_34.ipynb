```python
import matplotlib.pyplot as plt

tiempo_inicial=0
tiempo_final=5.22
velocidad_inicial=5

diferencia=0.01

numero_particiones=int((tiempo_final-tiempo_inicial)/diferencia)
lista_tiempo_x=[tiempo_inicial]
lista_velocidad_y=[velocidad_inicial]

def funcion_para_g(v):
    if v>1.5:
        ecuacion = -9.8*(0.125*((v/1.5)**(-1/2)))
        return ecuacion
    else:
        ecuacion = -9.8*(0.1+(0.2-0.1)/(1+2*v))
        return ecuacion


for i in range (numero_particiones):
    velocidad_inicial = velocidad_inicial + diferencia*funcion_para_g(velocidad_inicial)
    lista_velocidad_y.append(velocidad_inicial)
    
    
    tiempo_inicial = tiempo_inicial + diferencia
    lista_tiempo_x.append(tiempo_inicial)
    
fig, gx = plt.subplots()
gx.plot(lista_tiempo_x, lista_velocidad_y)
gx.set(xlabel='Tiempo (s)', ylabel='Velocidad (m/s)',
       title='Gráfica de la situación')
gx.grid()
```
